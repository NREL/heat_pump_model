```python
##### Validating to Kosmodakais
hp_kos = heat_pump()

# Setting COP performance
hp_kos.hot_temperature_desired = 150.0
hp_kos.cold_temperature_available = 80
hp_kos.cold_buffer = 0
hp_kos.hot_buffer = 0
hp_kos.carnot_efficiency_factor_flag = True
hp_kos.carnot_efficiency_factor = 0.41
#hp_corberan.refrigerant_flag = False
#hp_corberan.refrigerant = 'R1234ze(Z)'

hp_kos.calculate_COP()
print('')

hp_kos.process_heat_requirement = [0, 0, 0, 0.1365, 0.6824, 0.6824, 0.6824, 0.6824, 0.6824, 
    0.6824, 0.6824, 0.6824, 0.6824, 0.6824, 0.6824, 0.6824, 0.6824, 0.6824, 0.6824, 
    0.6824, 0.6824, 0.6824, 0.6824, 0, 0, 0] * 364 
# Getting the heating load
hp_kos.calculate_energy_and_mass_flow()

## Keeping in Euros
hp_kos.capital_cost_per_size = 150
hp_kos.gas_fixed_o_and_m_per_size = 0.04*hp_kos.capital_cost_per_size
hp_kos.utility_rate_kwh = [0.07]*8760
hp_kos.utility_rate_kw = 0.0
hp_kos.variable_o_and_m_per_mmbtu = 0.00
print('')
hp_kos.calculate_heat_pump_costs()

hp_kos.existing_gas = True
hp_kos.gas_capital_cost_per_size = 0.0
hp_kos.gas_fixed_o_and_m_per_size = 0.0
hp_kos.gas_variable_o_and_m_per_mmbtu = 0.01 # $/MMBTU
hp_kos.gas_price_MMBTU = [10.58] * 8760 
hp_kos.gas_efficiency = 0.9

print('')
hp_kos.calculate_natural_gas_comparison()

print('')
hp_kos.calculate_cash_flow()
#print('')
#print(sum(hp_kos.capacity_factor)/8760)
```

```python
## Setting up a contour plot of prices
x = [*range(20, 120, 1)]
y = [*range(10, 80, 1)]
for i in range(0, len(x)):
    x[i] = x[i]/10
for i in range(0, len(y)):
    y[i] = y[i]/1000

#array_PBP = np.zeros((len(x),len(y)))
print(x)
print(y)
#print(array_PBP)
x_solved = []
y_solved = []
z_solved = []

for i in range(0, len(x)):
    for j in range(0, len(y)):
        hp_kos.gas_price_MMBTU = [x[i]]*8760 
        hp_kos.utility_rate_kwh = [y[j]]*8760

        hp_kos.calculate_heat_pump_costs()
        hp_kos.calculate_natural_gas_comparison()
        try:
            hp_kos.calculate_cash_flow()
            if hp_kos.payback_period > 0.0 and hp_kos.payback_period < 20.0:
                x_solved.append(x[i])
                y_solved.append(y[j])
                z_solved.append(hp_kos.payback_period)
        except:
            continue

        print(i, j)
print(x_solved)
print(y_solved)
print(z_solved)
```

```python
fig = go.Figure(data=go.Scatter(
    x = x_solved,
    y = y_solved,
    mode='markers',
    marker=dict(
        size=16,
        color=z_solved, #set color equal to a variable
        colorscale='Jet', # one of plotly colorscales
        showscale=True
    )
))
fig.update_traces(mode='markers', marker_line_width=0, marker_size=9)
fig.update_xaxes(title_text = 'Gas Price (€/MMBTU)')
fig.update_yaxes(title_text = 'Electricity Price (€/kWh)')
fig.update_layout(template = 'simple_white', width = 600, height = 800)
fig.show()
```

```python
df_output = pd.DataFrame(0, columns = ['x', 'y', 'z'], index = range(0,len(x_solved)))
df_output['x'] = x_solved
df_output['y'] = y_solved
df_output['z'] = z_solved
df_output.to_excel('case_study_break_even.xlsx')


#print(hp_kos.process_heat_requirement_kw)

'''hp_kos.utility_rate_kwh = [0.02]*8760
hp_kos.calculate_heat_pump_costs()
print('')
#hp_kos.utility_rate_kwh = [0.04]*8760
#hp_kos.calculate_heat_pump_costs()
print('')
#hp_kos.utility_rate_kwh = [0.05]*8760
#hp_kos.calculate_heat_pump_costs()
print('')
hp_kos.utility_rate_kwh = [1.0]*8760
hp_kos.calculate_heat_pump_costs()
print('')'''
```
